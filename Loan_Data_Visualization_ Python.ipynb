# Loan Data Visualization
This dataset consists of more than 9,500 loans with information on the loan structure, the borrower, and whether the loan was pain back in full. This data was extracted from LendingClub.com, which is a company that connects borrowers with investors. 

```python
# Load packages
import pandas as pd
import seaborn as sns
df = pd.read_csv('loan_data.csv')
```

## Cleaning the Dataset

```python
#changing columns names
df = df.rename(columns = {'credit.policy':'credit_policy', 'int.rate':'int_rate',
                          'log.annual.inc':'log_annual_inc','days.with.cr.line':'days_with_cr_line',
                          'revol.bal':'revol_bal', 'revol.util':'revol_util','inq.lats.6mths':'inq_last_6mths',
                          'delinq.2yrs':'delinq_2yrs', 'pub.rec':'pub_rec', 'not.fully.paid':'not_fully_paid'})
df.dtypes
```

# Data Visualization

## How many loans were taken for each category?



```python
sns.set(rc = {'figure.figsize':(10,7)})
ax = sns.countplot(y = df.purpose, order = df['purpose'].value_counts().index)
ax.set(title="Number of Loans per Category", xlabel ='Number of Loans', ylabel ='Purpose');
```

## How many loans were fully paid?

```python
sns.set(rc = {'figure.figsize':(9,3)})
ax = sns.countplot( df.not_fully_paid)
ax.set(title="Number of Loans Fully Paid", xlabel ='Fully Paid', ylabel ='Number of loans')
ax.legend(['YES','NO']);
```

## How long does it take for a loan to be fully paid?


```python
df_paid = df.query('not_fully_paid == 0')

sns.set(rc = {'figure.figsize':(15,3)})
sns.boxplot(x=df_paid.days_with_cr_line/365).set(title="Average Payment Time", xlabel ='Years');
```

## How long does it take for a loan to be fully paid for each purpose?

```python
sns.set(rc = {'figure.figsize':(15,7)})
sns.boxplot(y=df_paid.days_with_cr_line/365,x =df_paid.purpose ).set(title="Average Payment Time per Purpose", xlabel ='Years');
```
